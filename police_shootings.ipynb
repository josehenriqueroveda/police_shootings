Import libs

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from collections import Counter
```

Loading the data

```
df = pd.read_csv('/content/drive/My Drive/ColabNotebooks/Projects/PoliceKillings/PoliceKillingsUS.csv', encoding='ISO-8859-1')
df
```

**Counting the people killed by gender.**

```
df.gender.value_counts()
```

```
sns.set(style='darkgrid', font_scale=1.1)
sns.countplot(x=df.gender, palette='hot')
plt.title('People killed by gender')
```

**Visualizing if the victim was armed at the time and what type of weapon**

```
armed = df.armed.value_counts()
print(armed)
```

```
plt.figure(figsize=(14,7))
sns.barplot(x=armed[:7].index,y=armed[:7].values, palette='hot')
plt.ylabel('Count of people')
plt.xlabel('Weapon Type')
plt.title('Weapon carried',fontsize=15)
```

**Analysis of the age of the victims**

```
view_age =['Above 21' if i >= 21 else 'Below 21' for i in df.age]
dataframe = pd.DataFrame({'age':view_age})
sns.countplot(x=dataframe.age, palette='hot')
plt.ylabel('Number of Killed People')
plt.title('Age of killed people',fontsize=15)

print(f'Avarage age of victims: {df.age.mean()}')
```

**Racial classification of people killed**

```
sns.countplot(data=df, x='race', palette='hot')
plt.title('Race of killed people',fontsize=15)
```

**Cities and states where deaths occurred**

```
city = df.city.value_counts()
plt.figure(figsize=(14,7))
sns.barplot(x=city[:12].index,y=city[:12].values, palette='hot')
plt.xticks(rotation=45)
plt.title('Cities where deaths occurred',fontsize=15)
```

```
state = df.state.value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=state[:20].index,y=state[:20].values, palette='hot')
plt.title('States where deaths occurred',fontsize=15)
```

**Victims with mental illness**

```
sns.countplot(df.signs_of_mental_illness, palette='hot')
plt.xlabel('Mental illness')
plt.ylabel('Count of people')
plt.title('Victims with mental illness',fontsize = 15)
```

**Threat level and flee types**

```
sns.countplot(df.threat_level, palette='hot')
plt.xlabel('Threat level')
plt.title('Threat level of people killed', fontsize = 15)
```

```
sns.countplot(df.flee, palette='hot')
plt.xlabel('Flee Types')
plt.title('Flee types',fontsize = 15)
```

**Did the police have a body camera?**

```
sns.countplot(df.body_camera, palette='hot')
plt.xlabel('Having Body Cameras')
plt.title('Did the police have a body camera?',fontsize = 15)
```
